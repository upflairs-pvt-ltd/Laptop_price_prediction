### DATA TRANSFORMATION

```python
df['rating'] = df['rating'].map(lambda x:x[0])
```

```python
df['warranty'] = df['warranty'].map({'1 year':1,'No warranty':0,'2 years':2,'3 years':3})
```

```python
transform_columns = ['processor_gnrtn','ram_gb','ssd','hdd','graphic_card_gb']

for column in transform_columns:
    df[column] = df[column].str.extract('(\d+)', expand=False)
    df[column] = df[column].astype('float')
    df[column]  = df[column].fillna(df[column].mean())
    df[column] = df[column].astype('int')
    
```

```python
df.head()
```

### ONE HOT ENCODING

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder =  OneHotEncoder(drop='if_binary')
```

```python
one_hot_columns = ['brand','processor_brand','processor_name','ram_type','os','os_bit','weight','Touchscreen','msoffice']
one_hot_encoding_df = df[one_hot_columns]
one_hot_encoding_df.head()
```

```python
## ONEHOT ENCODED OUR CATEGORICAL COLUMNS
onehot_encoded = onehot_encoder.fit_transform(one_hot_encoding_df).toarray()
onehot_encoded
```

```python
encoded_column_names = onehot_encoder.get_feature_names_out(one_hot_encoding_df.columns)
encoded_column_names
```

```python
onehot_data = pd.DataFrame(onehot_encoded , columns=encoded_column_names)
onehot_data.head()
```

```python
## NOW WE HAVE TWO DATAFRAME (DF , ONEHOT_DATA), SO WE WILL COMBINE BOTH TO GET A SINGLE DATAFRAME

df.drop(one_hot_columns,axis=1,inplace=True)
df.head()
```

```python
onehot_data.head()
```

```python
## RESETING INDEXES OF COLUMNS

df.reset_index(drop=True,inplace=True)
onehot_data.reset_index(drop=True,inplace=True)
```

```python
## NOW WE HAVE COMPLETE DATAFRAME 

df = pd.concat([df,onehot_data],axis=1)
df.head()
```

```python
## ZERO NONE IN MY DATAFRAME 
df.isnull().sum()
```

```python
df['hdd'].dtypes
```

```python
## PROGRAM TO CHANGE THE DATA TYPE OF ALL COLUMNS 

for column in df.columns:
    if  df[column].dtypes != int:
        df[column] = df[column].astype('int')


```

```python
## NOW YOU CAN SEE ALL COLUMNS DATA TYPE ARE "INT32"
df.dtypes
```

```python
df.head()
```

```python
## SAVED ALL COLUMN NAMES 
all_column_name = list(df.columns)

joblib.dump(all_column_name,'models/column_names.lb')
all_column_name
```

### SPLITTING THE DATASET 

```python
X = df.drop('Price',axis=1)
Y = df[['Price']]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

### LinearRegression Training

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
print(lr.score(x_train,y_train))
print(lr.score(x_test,y_test))
```

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
rdf = RandomForestRegressor()

rdf.fit(x_train,y_train)
print(rdf.score(x_train,y_train))
print(rdf.score(x_test,y_test))
```

### AdaBoostRegressor  , GradientBoostingRegressor

```python
from sklearn.ensemble import AdaBoostRegressor , GradientBoostingRegressor
svm = GradientBoostingRegressor()

svm.fit(x_train,y_train)
print(svm.score(x_train,y_train))
print(svm.score(x_test,y_test))
```

### Xgboot Algorithm

```python
import xgboost as xgb

dtrain = xgb.DMatrix(x_train, label=y_train)
dtest = xgb.DMatrix(x_test, label=y_test)

# Define XGBoost parameters
params = {
    'objective': 'reg:squarederror',  # Regression task
    'eval_metric': 'rmse',  # Root Mean Squared Error
    'max_depth': 3,  # Maximum depth of a tree
    'eta': 0.1,  # Learning rate
    'subsample': 0.8,  # Fraction of samples used for training
    'colsample_bytree': 0.8  # Fraction of features used for training
}

# Train the XGBoost model
num_round = 100  # Number of boosting rounds
xgb_model = xgb.train(params, dtrain, num_round)

y_test_pred = xgb_model.predict(dtest)
y_train_pred = xgb_model.predict(dtrain)

from sklearn.metrics import r2_score
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)
print(r2_train ,r2_test )
```
